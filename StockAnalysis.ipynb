```python
def inc_dec(c,o):
    if c>o:
        value= "increase"
    elif c<o:
        value = "decrease"    
    else: 
        value = "same"
    return value

df["Status"]= [inc_dec(c, o) for c, o in zip(df.Close, df.Open)]
df
```

```python
df["Middle"] = (df.Open+df.Close)/2
df["height"]=  abs(df.Open-df.Close)
```

```python
p = figure(x_axis_type='datetime',width = 1000, height=300)
# p.title="candlestick chart"
p.grid.grid_line_alpha = 0.3

p.segment(df.index, df.High,df.index, df.Low, color="Black")
hours_12 = 12*60*60*1000
p.rect(df.index[df.Status=="increase"], df.Middle[df.Status== "increase"] ,hours_12,
                df.height[df.Status== "increase"], fill_color="green", line_color="black")

p.rect(df.index[df.Status=="decrease"], df.Middle[df.Status== "decrease"],hours_12,
       df.height[df.Status== "decrease"], fill_color="red", line_color="black")

output_file("Chart.html")
show(p)
```

```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show, output_file

start_time= datetime.datetime(2016,3,1)
end_time=datetime.datetime(2017,3,10)

df = data.DataReader(name="GOOG", data_source="yahoo", start = start_time, end= end_time)

df

def inc_dec(c,o):
    if c>o:
        value= "increase"
    elif c<o:
        value = "decrease"    
    else: 
        value = "same"
    return value

df["Status"]= [inc_dec(c, o) for c, o in zip(df.Close, df.Open)]
df

df["Middle"] = (df.Open+df.Close)/2
df["height"]=  abs(df.Open-df.Close)

p = figure(x_axis_type='datetime',width = 1000, height=300, responsive=True)
# p.title="candlestick chart"
p.grid.grid_line_alpha = 0.3

p.segment(df.index, df.High,df.index, df.Low, color="Black")

hours_12 = 12*60*60*1000
p.rect(df.index[df.Status=="increase"], df.Middle[df.Status== "increase"] ,hours_12,
                df.height[df.Status== "increase"], fill_color="#CCFFFF", line_color="black")



p.rect(df.index[df.Status=="decrease"], df.Middle[df.Status== "decrease"],hours_12,
       df.height[df.Status== "decrease"], fill_color="#FF3333", line_color="black")

output_file("Chart.html")
show(p)

```
